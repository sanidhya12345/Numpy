```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
df=pd.read_csv('cricket_players.csv')


df.head()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=df,x='Country',y='Runs')
plt.title("Country vs Total Runs")
plt.xlabel("Country")
plt.ylabel("Total Runs")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(data=df['Average'])
plt.show()
```

```python
plt.figure(figsize=(20,40))

xlabels=['Matches Played','Runs','Wickets','Average','Strike Rate','Hundres','Fifties']
ylabels=['Matches Played','Runs','Wickets','Average','Strike Rate','Hundres','Fifties']
sns.heatmap(data=df,annot=True,cmap='Blues',xticklabels=xlabels,yticklabels=ylabels)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df,x='Matches Played',y='Runs')
plt.show()
```
