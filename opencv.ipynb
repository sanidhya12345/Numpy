```python
import cv2
import matplotlib.pyplot as plt
```

```python
#imread read the image in BGR format

img=cv2.imread('chameleon.jpg') 

#convert the image into RGB format

img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#render a image 

plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off") #for removing the axis values 
plt.show()
print(img_rgb.shape)

#print the pixel value

pixel=img_rgb[500,300]
print(pixel)

'''
cv2.imread('tv.jpeg'): Reads image file tv.jpeg as a BGR image.
cv2.cvtColor(img, cv2.COLOR_BGR2RGB): Converts the BGR image to RGB for proper color display.
plt.imshow(img_rgb): Plots the image.
plt.axis('off'): Removes the axis.
img_rgb.shape: Prints image shape (height, width, channels).
img_rgb[500,380]: Retrieves pixel color at (500, 380) as an RGB triplet.
'''
```

```python
#convert image into Grayscale

img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray,cmap='gray')
plt.title('converted gray image')
plt.axis("off")
plt.show()

print(img_gray.shape)

pixel_gray=img_gray[500,300]
print(pixel_gray)

#how to change the name of image in cv2

cv2.imwrite("myimage.jpeg",img_gray)
```

## Canny Edge Detection

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

#how to read an image directly as a grayscale without converting it

img_grayscale=cv.cvtColor(img,cv.IMREAD_GRAYSCALE) #reads directly as a grayscale

assert img is not None, "file could not be read, check with os.path.exists()"

edges=cv.Canny(img_grayscale,150,200)

plt.subplot(121)
plt.imshow(img,cmap='gray')
plt.title("Original Image")
plt.xticks([])
plt.yticks([])

plt.subplot(122)
plt.imshow(edges,cmap='gray')
plt.title('Edge Image')
plt.xticks([])
plt.yticks([])

plt.show()

'''
In Matplotlib, ticks are the small marks along the axes that indicate data points or positions.

plt.xticks([]), plt.yticks([])

✅ What it means:

plt.xticks([]) → Removes all ticks (and tick labels) from the x‑axis.

plt.yticks([]) → Removes all ticks (and tick labels) from the y‑axis.

💡 Why do we do this?
When we’re showing images, we usually don’t need axis numbers because they make the image look cluttered.
So plt.xticks([]) and plt.yticks([]) give you a clean image without axis ticks.

'''
```
