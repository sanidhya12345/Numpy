## how to import the pyplot module from matplotlib library

```python
import matplotlib.pyplot as plt

#Imports the pyplot module from matplotlib as plt, which is a common alias used for plotting functions.
```

# Line Charts

```python
x=[1,21,3,41,5]
y=[-14,5,6,7,8]

plt.plot(x,y) 
plt.title("This is my first graph") 
plt.xlabel("Price of the sugar")
plt.ylabel("Qty of the sugar")

#title(),xlabel,ylabel are used to label the graph

#show method is used to renders the graph on the screen
plt.show()
```

```python
#Task: create on example with some meaninful data.


days=['Mon','Tue','Wed','Thu','Fri']
temperature=[22,21,23,24,22]

#'v-r' means: downward triangle markers (v), red line (r).

#plt.plot(days,temperature,'v-r') method 1
plt.plot(days,temperature,marker='v',color='r',linestyle='-') #method 2
plt.title("Temperature over days")
plt.xlabel("days")
plt.ylabel("temperature (°C)")

plt.show()
```

```python
months=['Jan','Feb','Mar','Apr','May','June']
sugar_price=[32,34,30,36,38,37]


plt.plot(months,sugar_price,marker='o',color='green',linestyle='-')
#the above line can be written like this: 
#plt.plot(months,sugar_price,'o-g')
plt.title("Sugar Price trend in 2025")
plt.xlabel("Month")
plt.ylabel("Price (₹ per kg)")
plt.grid(True)
plt.show()
```

```python
days=[1,2,3,4,5,6,7]
gold_prices=[8975,9923,8765,9115,8931,8769,8721]
plt.plot(days,gold_prices,marker='o',linestyle='--',color='gold',label='Gold Prices')
plt.title("Gold Market Trend")
plt.xlabel("Days")
plt.ylabel("Prices(USD)")
plt.legend() #legend is used to label and differentiate between multiple plots in the same figure.

plt.show()

```

```python

epochs = [1, 2, 3, 4, 5]
training_accuracy = [0.6, 0.72, 0.80, 0.85, 0.88]
validation_accuracy = [0.58, 0.70, 0.78, 0.82, 0.84]

plt.plot(epochs, training_accuracy, marker='o',linestyle='-', label='Training Accuracy')
plt.plot(epochs, validation_accuracy, marker='s',linestyle='--', label='Validation Accuracy')
plt.plot(training_accuracy, validation_accuracy,linestyle='--', marker='s', label=' Accuracy')
plt.title('Model Accuracy Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

# Bar Charts

```python
category=['Cars','Bikes','Trucks']
values=[78,67,45]

plt.bar(category,values)
plt.show()
```

```python
category=['Cars','Bikes','Trucks']
values=[78,67,45]

# add some different colors to the bar graph

plt.bar(category,values,color=['r','g','b'])
plt.title("Vehicles Sales")
plt.xlabel("Vehicle Category")
plt.ylabel("Sales")
plt.show()
```

```python
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [25, 40, 35, 50]

plt.style.use('dark_background')
plt.bar(products,sales,color='skyblue')
plt.title("Sales by Products")
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()
```

```python
cat = ['Car', 'Bike', 'Truck']
val = [78, 67, 45]

plt.bar(cat, val, color=['skyblue', 'orange', 'lightgreen'])
plt.style.use('ggplot')
plt.xlabel('Vehicle Type')
plt.ylabel('Count')
plt.title('Number of Vehicles by Type')

#the alpha parameter is used to set the transparency of the grid lines. 
plt.grid(axis='y', linestyle='--', alpha=0.8)

#same as previous but with the lighter grid
#plt.grid(axis='y', linestyle='--', alpha=0.2)
plt.show()
```

# Histogram

```python
import numpy as np

data=np.random.randn(100)
plt.hist(data,bins=10,edgecolor='green',color='red')
plt.title("Histogram")
plt.show()
```

```python
data = np.random.rand(1000)

plt.hist(data,bins=40,color='skyblue',edgecolor='black')

plt.title("Histogram")

plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
data=np.random.randn(1000)

#we are setting the figure size

plt.figure(figsize=(2,3))
plt.hist(data,bins=12,color="blue",alpha=0.8)
plt.legend(["Random Data"],loc="upper right")
plt.grid(axis='y',linestyle='--',alpha=0.2) # Gridlines on y-axis

plt.title("Histogram of Random Data", fontsize=14, fontweight="bold")
plt.xlabel("Value", fontsize=12)
plt.ylabel("Frequency", fontsize=12)

plt.show()
```

# Pie Chart

```python
departments = ['Cardiology', 'Neurology', 'Pediatrics', 'Orthopedics']
visits = [120, 90, 150, 100]

plt.pie(visits, labels=departments, autopct='%1.1f%%', startangle=140)
plt.title('Department-wise Patient Distribution')
plt.axis('equal')
plt.show()

# Add some styles to this plot
```

```python
departments = ['Cardio','Neuro','Pediatrics','Orthopedic']
patients = [120,80,150,90]
plt.pie(patients,labels=departments,startangle=90,colors=['skyblue','lightcoral','lightgreen','lightpink'],explode=(0.1,0,0,0))
plt.show()
```

```python
x=np.random.rand(5)
mylabel=['A','B','C','D','E']
myexplode=[0.2,0,0,0,0]
plt.pie(x, labels=mylabel, autopct='%1.1f%%', startangle=140, explode=myexplode, shadow=True)
plt.legend(title='Group A',loc="upper right")
plt.show()
```
