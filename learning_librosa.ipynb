```python
import matplotlib.pyplot as plt
import numpy as np
```

### Loading audio files using Librosa

```python
import librosa
import librosa.display

from IPython.display import Audio

y,sr=librosa.load(librosa.ex('trumpet'))
Audio(y,rate=sr)

#librosa.ex: gives us a built in examples
#load function is used to load the audio files
```

### Plotting sound waves using librosa

```python
plt.figure(figsize=(10,4))
librosa.display.waveshow(y,sr=sr)
plt.grid(True)
plt.show()
```

### Trimming of the silence audio

```python

y_trimmed,_=librosa.effects.trim(y,top_db=20)
plt.figure(figsize=(10,4))
librosa.display.waveshow(y_trimmed)
plt.grid(True)
plt.show()



'''
 Explanation:
librosa.effects.trim() removes silent sections from beginning and end.

top_db=20 sets the threshold: anything below 20 dB of the peak is considered silence.

Only a portion of the trimmed waveform is plotted to zoom into the signal.
'''
```

### Spectogram and Short Time Fourier Transform (STFT)

```python
D=librosa.amplitude_to_db(np.abs(librosa.stft(y)),ref=np.max)
plt.figure(figsize=(10,4))
librosa.display.specshow(D,sr=sr,x_axis="time",y_axis="hz")
plt.colorbar(format='%+2.0f dB')
plt.show()
```
